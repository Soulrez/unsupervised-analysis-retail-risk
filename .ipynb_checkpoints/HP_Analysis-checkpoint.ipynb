```python
import numpy as np
```

```python
hp_dict = np.load("hp_dict_2d.npy")
```

```python
print(hp_dict)
```

```python
new_dict = {}
for key in hp_dict.item():
    new_dict["28_"+key] = hp_dict.item().get(key)+1
    new_dict["27_"+key] = hp_dict.item().get(key)    
print(new_dict)
```

```python
mean = {}
# replace with range(3,29)
for i in [27, 28]:
    results = [v for k,v in new_dict.items() if k.startswith(str(i))]
    mean[i] = np.mean(np.array(results))
print(mean)
```
