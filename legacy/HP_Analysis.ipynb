```python
import numpy as np
```

```python
hp_dict = np.load("hp_dict_2d_fe.npy")
```

```python
new_dict = {}
for key in hp_dict.item():
    new_dict[key] = hp_dict.item().get(key)
print(new_dict)
```

```python
mean = {}
# replace with range(3,29)
for i in range(3,29):
    results = [v for k,v in new_dict.items() if k.startswith(str(i))]
    mean[i] = np.mean(np.array(results))
print(mean)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(mean.keys(), mean.values(), '-o')

for x, y in zip(mean.keys(), mean.values()):
    plt.annotate(
        x,
        xy=(x, y), xytext=(-20, 20),
        textcoords='offset points', ha='right', va='bottom',
        bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5))
```
